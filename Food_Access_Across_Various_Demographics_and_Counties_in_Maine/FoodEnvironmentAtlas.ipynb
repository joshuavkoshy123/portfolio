```python
#Three lines to make our compiler able to draw:
import sys
import matplotlib
matplotlib.use('TkAgg')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

food_data = pd.read_csv("SupplementalDataCounty.csv", header=0, sep=",")
autauga = food_data['FIPS'] == 1001
autauga_data = food_data[autauga]
baldwin = food_data['FIPS'] == 1003
baldwin_data = food_data[baldwin]
barbour = food_data['FIPS'] == 1005
barbour_data = food_data[barbour]

x1 = autauga_data["Variable_Code"]
y1 = autauga_data["Value"]
x2 = baldwin_data["Variable_Code"]
y2 = baldwin_data["Value"]
x3 = barbour_data["Variable_Code"]
y3 = barbour_data["Value"]

barWidth = 0.25
#fig = plt.subplots(figsize =(12, 8)) 

fig, ax = plt.subplots()

x_autauga = [x + 2010 - barWidth for x in range(len(y1))] 
x_baldwin = [x + 2010 for x in range(len(y2))] 
x_barbour = [x + 2010 + barWidth for x in range(len(y3))] 

# creating the bar plot
ax.bar(x_autauga, y1,
        barWidth, label = 'Autauga County', color = 'blue', edgecolor = 'black')
ax.bar(x_baldwin, y2, 
        barWidth, label = 'Baldwin County', color = 'red', edgecolor = 'black')
ax.bar(x_barbour, y3, 
        barWidth, label = 'Barbour County', color = 'purple', edgecolor = 'black')
ax.plot(x1 - barWidth, y1, color = 'black', marker = 'o')
ax.plot(x2, y2, color = 'black', marker = 'o')
ax.plot(x3 + barWidth, y3, color = 'black', marker = 'o')
ax.set_xlim(xmin=2008, xmax=2020)
ax.set_ylim(ymin=0, ymax=250000)

ax.set_xlabel("Census Year")
ax.set_ylabel("Population")
ax.set_title("Population per Year")
ax.legend()

plt.show()
```

```python
# Low Access Population per County in Maine

import sys
import matplotlib
matplotlib.use('TkAgg')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

food_data = pd.read_csv("StateAndCountyData.csv", header=0, sep=",")

low_access_2010_data = food_data.loc[(food_data['Variable_Code'] == 'LACCESS_POP10') & (food_data['State'] == 'ME')]
low_access_2015_data = food_data.loc[(food_data['Variable_Code'] == 'LACCESS_POP15') & (food_data['State'] == 'ME')]
low_access_percent_change = food_data.loc[(food_data['Variable_Code'] == 'PCH_LACCESS_POP_10_15') & (food_data['State'] == 'ME')]

x1 = []
y1 = []
y2 = []
pc = []
for x in range(len(low_access_2015_data["Value"])):
    #if (low_access_2015_data["Value"].values[x] > low_access_2010_data["Value"].values[x]):
        x1.append(low_access_2010_data["County"].values[x])
        y1.append(low_access_2010_data["Value"].values[x])
        y2.append(low_access_2015_data["Value"].values[x])    
        pc.append(low_access_percent_change["Value"].values[x])
    
barWidth = 0.25
#fig = plt.subplots(figsize =(12, 8)) 

fig, ax = plt.subplots()

x_low_access_2010 = [x - barWidth / 2 for x in range(len(y1))] 
x_low_access_2015 = [x + barWidth / 2 for x in range(len(y2))] 
x = [x for x in range(len(y1))]

# creating the bar plot
ax.barh(x_low_access_2010, y1,
        barWidth, label = '2010', color = 'blue', edgecolor = 'black')
ax.barh(x_low_access_2015, y2,
        barWidth, label = '2015', color = 'red', edgecolor = 'black')

for i in range(len(pc)):
    if (pc[i] > 0):
        plt.text(y2[i] + 5000, x[i], f"% change : {pc[i]:.2f}%", ha = 'center', Bbox = dict(facecolor = 'darkred', alpha =.8))
    else:
        plt.text(y1[i] + 5000, x[i], f"% change : {pc[i]:.2f}%", ha = 'center', Bbox = dict(facecolor = 'limegreen', alpha =.8))
#ax.plot(y1, x1, label = '2010', color = 'blue', marker = 'o')
#ax.plot(y2, x1, label = '2015', color = 'red', marker = 'o')

ax.set_ylabel("County")
ax.set_xlabel("Population")
ax.set_title("Low Access Population per County in Maine")
ax.invert_yaxis()
plt.yticks(x, x1)
ax.legend()

plt.show()
```

```python
# Low Access Demographic Population in Maine

import sys
import matplotlib
matplotlib.use('TkAgg')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

food_data = pd.read_csv("StateAndCountyData.csv", header=0, sep=",")

low_access_white_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_WHITE15') & (food_data['State'] == 'ME')]
low_access_black_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_BLACK15') & (food_data['State'] == 'ME')]
low_access_hispanic_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_HISP15') & (food_data['State'] == 'ME')]
low_access_asian_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_NHASIAN15') & (food_data['State'] == 'ME')]
low_access_native_american_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_NHNA15') & (food_data['State'] == 'ME')]
low_access_pacific_islander_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_NHPI15') & (food_data['State'] == 'ME')]
low_access_multiracial_data = food_data.loc[(food_data['Variable_Code'] == 'PCT_LACCESS_MULTIR15') & (food_data['State'] == 'ME')]

sum_white = 0
sum_black = 0
sum_hispanic = 0
sum_asian = 0
sum_na = 0
sum_pi = 0
sum_mr = 0
for x in range(len(low_access_white_data["Value"])):
    #if (low_access_2015_data["Value"].values[x] > low_access_2010_data["Value"].values[x]):
        sum_white += low_access_white_data["Value"].values[x]
        sum_black += low_access_black_data["Value"].values[x]
        sum_hispanic += low_access_hispanic_data["Value"].values[x]
        sum_asian += low_access_asian_data["Value"].values[x]
        sum_na += low_access_native_american_data["Value"].values[x]
        sum_pi += low_access_pacific_islander_data["Value"].values[x]
        sum_mr += low_access_multiracial_data["Value"].values[x]
        
x = ['White', 'Black', 'Hispanic', 'Asian', 'Native American', 'Pacific Islander', 'Multiracial']
y = [sum_white, sum_black, sum_hispanic, sum_asian, sum_na, sum_pi, sum_mr]
barWidth = 0.25
#fig = plt.subplots(figsize =(12, 8)) 

fig, ax = plt.subplots()

# creating the bar plot
ax.bar(x, y, barWidth, color = 'blue', edgecolor = 'black')

for i in range(len(y)):
#    if (pc[i] > 0):
        plt.text(i, y[i] + 5000, round(y[i]), ha = 'center', Bbox = dict(facecolor = 'white', alpha =.8))
#    else:
#        plt.text(y1[i] + 750, x[i], f"% change : {pc[i]:.2f}%", ha = 'center', Bbox = dict(facecolor = 'limegreen', alpha =.8))
#ax.plot(y1, x1, label = '2010', color = 'blue', marker = 'o')
#ax.plot(y2, x1, label = '2015', color = 'red', marker = 'o')

ax.set_xlabel("Ethnicity")
ax.set_ylabel("Population")
ax.set_title("Low Access Demographic Population in Maine")
#ax.invert_yaxis()
plt.xticks(x)
#ax.legend()

plt.show()
```
